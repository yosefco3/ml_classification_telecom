# Final Project 

The telecom operator Interconnect would like to be able to forecast their churn of clients. If it's discovered that a user is planning to leave, they will be offered promotional codes and special plan options. Interconnect's marketing team has collected some of their clientele's personal data, including information about their plans and contracts.

### Interconnect's services

Interconnect mainly provides two types of services:

1. Landline communication. The telephone can be connected to several lines simultaneously.
2. Internet. The network can be set up via a telephone line (DSL, *digital subscriber line*) or through a fiber optic cable.

Some other services the company provides include:

- Internet security: antivirus software (*DeviceProtection*) and a malicious website blocker (*OnlineSecurity*)
- A dedicated technical support line (*TechSupport*)
- Cloud file storage and data backup (*OnlineBackup*)
- TV streaming (*StreamingTV*) and a movie directory (*StreamingMovies*)

The clients can choose either a monthly payment or sign a 1- or 2-year contract. They can use various payment methods and receive an electronic invoice after a transaction.

### Data Description

The data consists of files obtained from different sources:

- `contract.csv` — contract information
- `personal.csv` — the client's personal data
- `internet.csv` — information about Internet services
- `phone.csv` — information about telephone services

In each file, the column `customerID` contains a unique code assigned to each client.

The contract information is valid as of February 1, 2020.

Target feature: the `'EndDate'` column equals `'No'`.

Primary metric: AUC-ROC.

Additional metric: Accuracy.

Assessment criteria:

- AUC-ROC < 0.75 — 0 SP
- 0.75 ≤ AUC-ROC < 0.81 — 4 SP
- 0.81 ≤ AUC-ROC < 0.85 — 4.5 SP
- 0.85 ≤ AUC-ROC < 0.87 — 5 SP
- 0.87 ≤ AUC-ROC < 0.88 — 5.5 SP
- AUC-ROC ≥ 0.88 — 6 SP

### Load data and EDA

```python
data = {}
for df in ["personal","contract","phone","internet"]:
    data[df] = pd.read_csv(f"/datasets/final_provider/{df}.csv")

for df in data:
    data[df] = data[df].applymap(lambda s: s.lower() if type(s) ==
                 str else s, na_action='ignore')
    data[df].columns = [x.lower() for x in data[df].columns]
    display(f"df name: {df.upper()}")
    display(f"shape: {data[df].shape}")
    display(data[df].sample(10))
    display("***************************************************************************************")
```

```python
for df in data:
    data[df].info()
    display("***********************************************************")
```

* the reason is probably that some of the client have both phone and internet and some not.

```python
for df in data:
    print(f"{df} Duplicated: {data[df].duplicated().sum()}")
```

* lets see the target column:

```python
data["contract"]["enddate"].value_counts()
```

* lets join all the dataframe into one:

```python
df = data['personal'].merge(data['contract'], on  ="customerid")
df.head()
```

* we will create a churn column:

```python
df["churn"] = (df["enddate"]!="no").astype('int')
```

```python
df['churn'].value_counts(normalize=True)
```

```python
sns.countplot(data = df,x = 'churn');
```

* we have here unbalanced data - 73% to 27%.

* now we will add the "phone" table. first we remember that not all customers have phone, so we will add it in left join, and we will use 0,1,2 values: 
    * 0 - someone hasnt phone
    * 1- have phone but not multiline
    * 2 - have multiline
    

```python
df = df.merge(data["phone"], how = 'left', on = "customerid")
```

```python
df.rename(columns={"multiplelines":"phone"},inplace=True)
df.head()
```

```python
df = df.merge(data["internet"], how = 'left', on = "customerid")
df.head()
```

* we will add a "time_length" column to measure the days the customer is in the company. if there isn't end date, we will subtract the begining time from now (1th Feb 2020)

```python
def length(r):
    if r["enddate"]=='no':
        return pd.to_datetime('2020-02-01') - pd.to_datetime(r['begindate']) 
    else:
        return pd.to_datetime(r['enddate']) - pd.to_datetime(r['begindate']) 

df['time_length'] = df.apply(length,axis=1).dt.days
```

```python
df.head()
```

* we will drop now coulmns we dont need for training a ML model:

```python
df = df.drop(columns=['customerid','begindate','enddate'])
```

* if we try to convert "total charges" to numeric, we get an error. lets see how much non-numeric-convertables values are:

```python
pd.to_numeric(df["totalcharges"], errors='coerce').isna().sum()
```

* it's not a big amount so we will drop them:
* ( the errors='coerce' make this values nulls:)

```python
df["totalcharges"] = pd.to_numeric(df["totalcharges"], errors='coerce')
```

```python
df.isna().sum()
```

* the other nulls are in the categorical columns , we will treat them later in differnt way.

```python
df.dropna(subset=['totalcharges'], how='all', inplace=True)
```

```python
df["totalcharges"] = df["totalcharges"].astype('float')
```

```python
df.info()
```

* we have some ordinal columns: like phone (no-phone, one line, multi lines) , type and internet service, we would treat them now.
* all these columns are ordinals variables, because one service is expensive from the other.
* we will replace the nulls with 0, and order the other values.

```python
def fix_phone(x):
    if pd.isnull(x):
        x=0
    elif x=='no':
        x=1
    elif x=='yes':
        x=2
    return x

df["phone"] = df["phone"].apply(fix_phone)
```

```python
df["internetservice"].value_counts()
```

```python
def fix_internet(x):
    if pd.isnull(x):
        x=0
    elif x=='dsl':
        x=1
    elif x=='fiber optic':
        x=2
    return x

df["internetservice"] = df["internetservice"].apply(fix_internet)
```

```python
df["type"].value_counts()
```

```python
def fix_type(x):
    if x=="month-to-month":
        x=0
    elif x=='one year':
        x=1
    elif x=='two year':
        x=2
    return x

df["type"] = df["type"].apply(fix_type)
```

```python
df.info()
```

* now we will fix all the internet features:
* we will replace the nulls and the "no" values with 0, and yes with 1.

```python
def fix_internet_features(x):
    if pd.isnull(x) or x=='no':
        x = 0
    else:
        x=1
    return x
    

for col in ['onlinesecurity','onlinebackup',
            'deviceprotection','techsupport','streamingtv','streamingmovies',
            'paperlessbilling','partner','dependents']:
    df[col] = df[col].apply(fix_internet_features)
    
df.head()
```

* there remains only two categorical columns, we will do OHE to make them numerical:

```python
df = pd.concat([df.drop(columns=['gender','paymentmethod']),pd.get_dummies(df[['gender','paymentmethod']],drop_first=True)],axis=1)
```

```python
df.head()
```

### devide the data to train test, and treat imbalancing:

```python
X= df.drop('churn', axis=1)
y= df['churn']
```

*we will use SMOTEENN from imblearn. the description is:
        "Over-sampling using SMOTE and cleaning using ENN.
        Combine over- and under-sampling using SMOTE and Edited Nearest Neighbours."

```python
oversample = SMOTEENN(random_state=42)
X, y = oversample.fit_resample(X, y)
```

```python
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
scalar = MinMaxScaler()
X_train_scaled  = scalar.fit_transform(X_train)
X_test_scaled = scalar.transform(X_test)
```

## Models

### Random forest

```python
# we will use this params grid below too.
n_estimators =[64,84,110,144]
max_depth=[5,6,7,8,9,10]
random_state=[42]
param_grid={'n_estimators':n_estimators ,'max_depth':max_depth,'random_state':random_state}
```

```python
rfc = RandomForestClassifier(random_state=42,class_weight="balanced")
forest_grid = GridSearchCV(rfc,param_grid,scoring='roc_auc')
```

```python
forest_grid.fit(X_train_scaled,y_train)
```

```python
forest_grid.best_params_
```

```python
y_pred= forest_grid.predict(X_test_scaled)
```

```python
d={
        "accuracy":[],
        "precision":[],
        "recall":[],
        "f1_score":[],
        "AUC":[]
    }
models_names =[]
models =[]

def check(model,x_test,y_test,model_name,d,models_names):
    preds = model.predict(x_test)
    d["accuracy"].append(accuracy_score(y_test,preds))
    d["precision"].append(precision_score(y_test,preds))
    d["recall"].append(recall_score(y_test,preds))
    d["f1_score"].append(f1_score(y_test,preds))
    d["AUC"].append(roc_auc_score(y_test, preds))
    models_names.append(model_name)
    models.append(model)
    d = pd.DataFrame(d,index=models_names).sort_values(["f1_score"],ascending=False)
    display(d)
#     print(classification_report(y_test,preds))
    fig,axes = plt.subplots(nrows=1, ncols=2,figsize=(8,2),dpi=150);
    fig.suptitle('Confusion matrix')
    plot_confusion_matrix(model,x_test,y_test,ax=axes[0]);
    plot_confusion_matrix(model,x_test,y_test,normalize='all',ax=axes[1]);
```

```python
check(forest_grid,X_test_scaled,y_test,"random_forest",d,models_names)
```

```python
def plot_roc():
    fig, axes = plt.subplots(nrows=1, ncols=1,figsize=[10,4],dpi=200)
    axes.set_title("ROC curve for the models")
    for i in range(len(models)):
        plot_roc_curve(models[i],X_test,y_test,ax=axes,label=f"{models_names[i]}, AUC={d['AUC'][i]:.2f}");

plot_roc()
```
